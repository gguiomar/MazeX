# environments

```python
class world(object):
    def __init__(self):
        return

    def get_outcome(self):
        return
    
    def get_all_outcomes(self):
        
        outcomes = {}
        for state in range(self.n_states):
            for action in range(1 if self.n_actions == 0 else self.n_actions):
                next_state, reward = self.get_outcome(state, action)
                outcomes[state, action] = [(1, next_state, reward)]
        return outcomes

class n_armed_bandit():
    """
    World: N-Armed bandit.
    Only one state, multiple actions.
    Each action returns different amount of reward.
    For each action, rewards are randomly sampled from normal distribution, 
        with a mean associated with that arm and unit variance.
    """
    
    def __init__(self, arm_number):
        self.name = "n_armed_bandit"
        self.n_states = 1
        self.n_actions = arm_number
        self.dim_x = 1
        self.dim_y = 1
        
        self.mu = [np.random.normal(0,1) for a in range(self.n_actions)]
        
    def get_outcome(self, state, action):
        
        self.rewards = [np.random.normal(self.mu[i],1) for i in range(self.n_actions)]
        next_state = None
        
        reward = self.rewards[action]
        return int(next_state) if next_state is not None else None, reward
    
class drifting_n_armed_bandit():
    """
    World: N-Armed bandit.
    Only one state, multiple actions.
    Each action returns different amounts of rewards.
    For each action, rewards are randomly sampled from normal distribution, 
        with a mean associated with that arm and unit variance.
    In the case of the non-stationary bandit, the mean reward associated with each arm 
        follows a Gaussian random.
    """

    def __init__(self, arm_number, drift):
        self.name = "drifting_n_armed_bandit"
        self.n_states = 1
        self.n_actions = arm_number
        self.dim_x = 1
        self.dim_y = 1
        
        self.mu_min = 0.25
        self.mu_max = 0.75
        self.drift = drift
        
        # self.mu = [np.random.uniform(self.mu_min, self.mu_max) for a in range(self.n_actions)]
        self.mu = [np.random.normal(0,0) for a in range(self.n_actions)]

    def update_mu(self):
        self.mu += np.random.normal(0, self.drift, self.n_actions)
        # self.mu[self.mu > self.mu_max] = self.mu_max
        # self.mu[self.mu < self.mu_min] = self.mu_min
            
    def get_outcome(self, state, action):
        
        self.update_mu()
        self.rewards = [np.random.normal(self.mu[i],1) for i in range(self.n_actions)]
        next_state = None
        
        reward = self.rewards[action]
        return int(next_state) if next_state is not None else None, reward
```

# pathway bandit

### functions

```python
class signal_armed_bandit():
    
    def __init__(self, n_arms):
        self.name = "n_armed_bandit"
        self.n_states = 1
        self.n_actions = n_arms
        
    def get_outcome(self, action, rwd_sig, rwd_mag):
        
        self.mu = np.zeros(self.n_actions)
        self.mu[rwd_sig] = rwd_mag
        self.rewards = [np.random.normal(self.mu[i],1) for i in range(self.n_actions)]
        
        return self.rewards[action]

def e_greedy_policy(q, epsilon):
    if np.random.uniform(0,1) > epsilon:
        return np.argmax(q)
    else:
        # choose randomly from all but the highest q value
        a_list = np.arange(len(q))
        arg_max = np.argmax(q)
        choices = np.delete(a_list, arg_max)
        return np.random.choice(np.delete(a_list, arg_max))

def softmax_policy(q, beta):
    axis = None
    x = q - q.max(axis=axis, keepdims=True)
    y = np.exp(x * beta)

    for i, e in enumerate(y):
        if np.isinf(e):
            y[i] = 1

    action_prob = y / y.sum(axis=axis, keepdims=True)
    selected_action = np.argmax(np.random.multinomial(1, action_prob, size=1))

    return selected_action

def greedy_policy(q):
    return np.argmax(q)

def moving_average(x, w):
    return np.convolve(x, np.ones(w), 'valid') / w

def tfp(p, rpe):
    if p == 0:
        if rpe > 0:
            return rpe
        else:
            return 0
    if p == 1:
        if rpe < 0:
            return -rpe
        else:
            return 0
        
def nl_tfp(p, rpe):
    if p == 0:
        if rpe > 0:
            return rpe
        else:
            return 0
    if p == 1:
        if rpe < 0:
            return rpe
        else:
            return 0

# integer triangular wave generator
def triangle_signal_generator(n_arms, n_cycles, n_steps):
    sig_range = n_arms * n_cycles
    csig = np.zeros(sig_range)
    t1 = 0
    t2 = 0
    c = 0
    n_steps = 10
    for t in range(sig_range-1):
        if csig[t] < n_arms-1 and t1 == 0:
            c = 1
            #print('a')
        if csig[t] == n_arms-1:
            c = -1
            t1 = 1
            #print('b')
        if csig[t] == 0 and t1 == 1:
            c = 1
            t1 = 0
            #print('c')
        csig[t+1] = csig[t]+c
        #print(csig[t])
    #plt.plot(csig)
    csig = csig.astype(int)

    nsteps = 10
    csig_h = []
    for t in range(n_arms * n_cycles):
        for st in range(n_steps):
            csig_h.append(csig[t])
            
    return csig, csig_h
```

# Bandit simulations

```python
# symmetric case

n_arms = 5
n_steps = 1000
epsilon = 0.5
min_alpha = 0.01
alpha = np.zeros(n_arms) + min_alpha
rwd_mag = 10

sab = signal_armed_bandit(n_arms)
n_pathways = 2

q = np.zeros((n_pathways, n_arms))
Q = np.zeros(n_arms)
rwd = np.zeros(n_steps)
q_ev= np.zeros((n_arms, n_steps*n_arms)) # evolution of q
q_ev_p= np.zeros((n_pathways, n_arms, n_steps*n_arms)) # evolution of q
a_hist = np.zeros(n_steps*n_arms) # action history

action = np.random.choice(np.arange(len(q)))
n_visits = np.zeros(n_arms) + min_alpha

beta_d = 1
beta_i = 1
temp = 1
c = 0
sig_h = np.zeros(n_arms * n_steps)
cos_sig = np.cos(np.linspace(0,n_arms,n_arms))+1
for sig in range(n_arms):
    for s in range(n_steps):
        # signal history
        sig_h[c] = sig
        
        # action selection
        action = softmax_policy(Q, temp)
        n_visits[action] += 1
        alpha[action] = 1/n_visits[action]
        
        # calculating and saving reward
        reward = sab.get_outcome(action, sig, rwd_mag)

        # rpe calc
        rpe_Q = reward - Q[action]
        
        # action-value update rule
        for p in range(n_pathways):
            a = action
            #print(sig, p, a, rpe_Q, tfp(p,rpe_Q))
            q[p, a] = q[p,a] + alpha[a] * tfp(p, rpe_Q)
            q_ev_p[p,action,c] = q[p, action]
        
        # joining action-values
        Q[action] = beta_d * q[0,action] + beta_i * q[1,action]
        q_ev[:,c] = Q
        a_hist[c] = action
        c+=1

fig, ax = plt.subplots(2, 2, figsize = (12,7))

for a in range(n_arms):
    ax[0,0].plot(q_ev[a,:])

ax[0,1].plot(moving_average(a_hist,200), 'k')
ax[0,1].plot(sig_h, 'k--')

for p in range(n_pathways):
    for a in range(n_arms):
        ax[1,p].plot(q_ev_p[p, a,:], label = str(a))
```

## triangular wave autoencoder

In this case we only perform a split in the RPEs, the bandits are still idependent agents with update equations given by

$$q(a_k)^p_{t+1} = q(a_k)^p_t + \alpha_p  f^p(R_t - Q(a_k))$$

and the global action-value function is given by

$$Q(a_k) = \sum_p \eta_p q^p(a_k)$$

At each timestep $t$ we sample an action $a_i^t$ from the distribution $\pi(Q|t)$ forming a sequence of actions through time $\tilde a(t)$. We then decode the signal through a simple moving average of the sequence of sampled actions via

$$x(t) = \int_{-\infty}^{\infty} k^{\omega}(u)\tilde{a}(u)du$$

where $k^\omega(u)$ is a square kernel of size $\omega$

```python
n_arms = 10
n_steps = 300
epsilon = 0.5
min_alpha = 0.01
alpha = np.zeros(n_arms) + min_alpha
alpha_c = 0.01
rwd_mag = 5

sab = signal_armed_bandit(n_arms)
n_pathways = 2

q = np.zeros((n_pathways, n_arms))
Q = np.zeros(n_arms)
rwd = np.zeros(n_steps)
q_ev= np.zeros((n_arms, n_cycles * n_steps * n_arms)) # evolution of q
q_ev_p= np.zeros((n_pathways, n_arms, n_cycles * n_steps * n_arms)) # evolution of q
a_hist = np.zeros(n_cycles * n_steps * n_arms) # action history

action = np.random.choice(np.arange(len(q)))
n_visits = np.zeros(n_arms) + min_alpha

beta_d = 1
beta_i = -5
temp = 2
c = 0
sig_h = []
rpe_Q_h = []

n_cycles = 10
csig = triangle_signal_generator(n_arms, n_cycles, n_steps)
act_cross = []

lamb = 0.98

for sig in csig[0]:
    act_cross.append(c)
    for s in range(n_steps):
            
        # signal history
        sig_h.append(sig)
        
        # action selection
        action = softmax_policy(Q, temp)
        n_visits[action] += 1
        alpha[action] = 1/n_visits[action]
        
        # calculating and saving reward
        reward = sab.get_outcome(action, sig, rwd_mag)

        # rpe calc
        rpe_Q = reward - Q[action]
        rpe_Q_h.append(rpe_Q)
        
        # action-value update rule
        for p in range(n_pathways):
            a = action
            # implement distance metric!
            q[p, a] = lamb * q[p,a] + alpha_c * tfp(p, rpe_Q)
            #q[p, a] = q[p,a] + alpha_c * tfp(p, rpe_Q)
            q_ev_p[p,action,c] = q[p, action]
        
        # joining action-values
        Q[action] = beta_d * q[0,action] + beta_i * q[1,action]
        q_ev[:,c] = Q[:]
        a_hist[c] = action
        c+=1
```

* Because the reward signal is the same all the action values will converge to the same value given enough iterations.
* The system is not able to pick up the signal and "fire" the correct action. Noise increases and the probability of choosing a given action is uniform.
* In other words, the entropy of the system increases! Exactly what we were after.
* If were able to now design a set of learning rules that decrease the entropy the signal will be able to be noticed again and the actions will be able to be selected.
* Complex cell filters are a way of reducing the entropy of a system due to policy compression.

# implementing the update equations with $g(a_k-a_i)$

current update rules, if a given action $a_i$ is taken at time $t$, the update rule for all actions is given by

$$q(a_k)^p_t = q(a_k)^p_t + \alpha_p  f^p(R_t - Q(a_k)) g^p(a_k - a_i)$$

```python
def gm(ak,aj,sig):
    #ak action taken, aj action being updated
    return np.exp(-np.power(ak - aj, 2.) / (2 * np.power(sig, 2.)))

def gm2(ak,aj,sig,kc):
    #ak action taken, aj action being updated
    return kc+np.exp(-np.power(ak - aj, 2.) / (2 * np.power(sig, 2.)))
```

```python
def plot_action_kernels(n_arms, kw, kg, kc):
    
    fig, ax = plt.subplots(1, 3, figsize = (15,3))
    aj = n_arms/2
    bf_d = np.asarray([kg[0] * gm2(aj,ak,kw[0],kc[0]) for ak in np.linspace(0,n_arms,n_arms)])
    bf_i = np.asarray([kg[1] * gm2(aj,ak,kw[1],kc[1]) for ak in np.linspace(0,n_arms,n_arms)])
    
    cm_bandits = np.zeros((n_arms, n_arms))
    for ai in range(n_arms):
        for aj in range(n_arms):
            cm_bandits[ai,aj] = kg[0] * gm2(ai,aj,kw[0],kc[0]) - kg[1] * gm2(ai,aj,kw[1],kc[1])
    
    ax[0].plot(bf_d, 'b')
    ax[0].plot(bf_i, 'r')
    ax[1].plot(bf_d - bf_i, 'k')
    cf = ax[2].imshow(cm_bandits, cmap='RdBu')
    fig.colorbar(cf, ax = ax[2])
    plt.show()
```

### asymmetries

### SR connectivity

We now assume that the connectivity between cagents $(i,j)$ defined by $\Omega_{ij}$ can be learned through the structure of the environment. We define this connectivty as a function of the transition probabilities between consequent actions i.e. how likely action $a_i$ is to be followed by action $a_j$. In order to simplify notation we define de sequence as $a\rightarrow a'$

$$\Omega_{ij} = M(a,a')$$

where

$$\Omega_{a,a'} \leftarrow \Omega_{a,a'} + \alpha( \mathbb{1} + \gamma \Omega_{a,a'}-\Omega_{a,a})$$ - watchout, this update rule needs some work in understanding

For convergence purposes the learning rate has a descaling factor $\lambda < 1$ for actions that are not executed

$$q(a_k)^p_t = \lambda q(a_k)^p_t + \alpha_p  f^p(R_t - Q(a_k)) $$

```python
# simulation parameters
n_arms = 10
n_steps = 500
min_alpha = 0.01
alpha = np.zeros(n_arms) + min_alpha
alpha_c = 0.01
alpha_m = 0.01
rwd_mag = 10 # rwd_mag should be part of the cost, we want to minimize the distance

# discount & eligibility
gamma_q = 0.97
gamma_m = 0.98
lambd_e = 0.99

# signal and environment
n_cycles = 5
sab = signal_armed_bandit(n_arms)
csig = triangle_signal_generator(n_arms, n_cycles, n_steps)
n_pathways = 2

# value functions
q = np.zeros((n_pathways, n_arms))
Q = np.zeros(n_arms)
rwd = np.zeros(n_steps)
q_ev= np.zeros((n_arms, n_cycles * n_steps * n_arms)) # evolution of q
q_ev_p= np.zeros((n_pathways, n_arms, n_cycles * n_steps * n_arms)) # evolution of q
a_hist = np.zeros(n_cycles * n_steps * n_arms) # action history
sig_h = []
rpe_Q_h = []

# learning parameters
action = np.random.choice(np.arange(len(q)))
n_visits = np.zeros(n_arms) + min_alpha
beta_d = 1
beta_i = 2
temp = 2
c = 0
a_p = 0 # previous action holder

# SR and eligibility
W = np.zeros((n_arms, n_arms))
Et = np.zeros(n_arms)

#plotting vars
act_cross = []

for sig in csig[0]:
    act_cross.append(c)
    for s in range(n_steps):
        
        # signal history
        sig_h.append(sig)
        
        # action selection
        a_k = softmax_policy(Q, temp) # current_action
        n_visits[action] += 1
        alpha[a_k] = 1/n_visits[a_k]
        
        # building SR representation with eligibility traces
        # a_k is next state, a_p previous state
        id_a = 0
        for i in range(n_arms):
            
            if i != a_p:
                Et[i] = lambd_e * Et[i]
            if i == a_p:
                Et[i] = lambd_e * Et[i] + 1
                
            for j in range(n_arms):
                if j == a_k:
                    id_a = 1
                else:
                    id_a = 0

                W[i,j] += alpha_m * (id_a + gamma_m * W[a_k,j] -  W[a_p,j]) * Et[i]
        
        # calculating and saving reward
        reward = sab.get_outcome(a_k, sig, rwd_mag)

        # rpe calc
        rpe_Q = reward - Q[a_k]
        rpe_Q_h.append(rpe_Q)
        
        # action-value update rule - old version with splitting
        for p in range(n_pathways):
            a_j = a_k
            q[p, a_j] = gamma_q * q[p,a_j] + alpha_c * tfp(p, rpe_Q)
            q_ev_p[p,a_j,c] = q[p, a_j]
        
        # joining action-values
        for a_j in range(n_arms):
            Q[a_j] = beta_d * q[0,a_j] + beta_i * q[1,a_j]

        q_ev[:,c] = Q
        a_hist[c] = a_k
        c+=1
        a_p = a_k # previous action
```

### random signal SR

```python
# integer triangular wave generator
def random_signal_generator(n_arms, n_cycles, n_steps):
    
    sig_range = n_arms * n_cycles
    csig = np.zeros(sig_range)
    
    for t in range(sig_range-1):
        c = np.random.choice([-3,-2,-1,0,1,2,3])
        csig[t+1] = csig[t]+c
        if csig[t+1] < 0:
            csig[t+1] = 0
        if csig[t+1] > n_arms-1:
            csig[t+1] = n_arms-1
            
    csig = csig.astype(int)
    csig_h = []
    for t in range(n_arms * n_cycles):
        for st in range(n_steps):
            csig_h.append(csig[t])
            
    return csig, csig_h
```

## bandit network with SR connectivity

```python
# main parameters 
n_arms = 10
n_steps = 500
epsilon = 0.5
min_alpha = 0.01
alpha = np.zeros(n_arms) + min_alpha
alpha_c = 0.01
alpha_o = 0.0001
alpha_m = 0.01
rwd_mag = 10 # rwd_mag should be part of the cost, we want to minimize the distance

# signal variables
sab = signal_armed_bandit(n_arms)
n_cycles = 3
csig = triangle_signal_generator(n_arms, n_cycles, n_steps)
act_cross = []

# defining value vectors
n_pathways = 2
q = np.zeros((n_pathways, n_arms))
Q = np.zeros(n_arms)
rwd = np.zeros(n_steps)
q_ev= np.zeros((n_arms, n_cycles * n_steps * n_arms)) # evolution of q
q_ev_p= np.zeros((n_pathways, n_arms, n_cycles * n_steps * n_arms)) # evolution of q
a_hist = np.zeros(n_cycles * n_steps * n_arms) # action history

action = np.random.choice(np.arange(len(q)))
n_visits = np.zeros(n_arms) + min_alpha

# policy parameters 
beta_d = 1
beta_i = 5
temp = 2
c = 0
sig_h = []
rpe_Q_h = []

# SR and eligibility traces
#W = np.zeros((n_arms, n_arms))
Et = np.zeros(n_arms)

gamma_q = 0.97
gamma_m = 0.98
lambd_e = 0.99

a_p = 0
a_k = 0
for sig in csig[0]:
    act_cross.append(c)
    for s in range(n_steps):
            
        # signal history
        sig_h.append(sig)
        
        # action selection
        a_p = a_k # past action
        a_k = softmax_policy(Q, temp)
        n_visits[a_k] += 1
        alpha[a_k] = 1/n_visits[a_k]
        
        # calculating and saving reward
        reward = sab.get_outcome(a_k, sig, rwd_mag)

        # rpe calc
        rpe_Q = reward - Q[a_k]
        rpe_Q_h.append(rpe_Q)
        
        # Successor representation connectivity update
        # a_k - new action
        # a_p - past action
        
        id_a = 0
        for i in range(n_arms):
            
            if i != a_p:
                Et[i] = lambd_e * Et[i]
            if i == a_p:
                Et[i] = lambd_e * Et[i] + 1
                
            for j in range(n_arms):
                if j == a_k:
                    id_a = 1
                else:
                    id_a = 0
                
                #print(i,j,a_k,a_p)
                #W[i,j] += alpha_m * (id_a + gamma_m * W[a_k,j] -  W[a_p,j]) * Et[i]
                
        # action-value update rule
        for p in range(n_pathways):
            for j in range(n_arms):
                if j == a_k:
                    q[p,j] += alpha_c * tfp(p, rpe_Q) #* (1+0.01 * W[j,a_k])
                else:
                    q[p,j] += alpha_o * W[j,a_k]
                q_ev_p[p,j,c] = q[p,j]

        # joining action-values
        Q[a_k] = beta_d * q[0,a_k] + beta_i * q[1,a_k]
        q_ev[:,c] = Q[:]
        a_hist[c] = a_k
        c+=1
```
